# PROBLEM STATEMENT:Classifying Iris Species

```python
import pandas as pd
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
iris_dataset
```

```python
y=iris_dataset['target']
y
```

```python
x=iris_dataset['data']
x
```

# We define the training and the test sets for building our machine learning model.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from pandas.plotting import scatter_matrix
```

```python
import seaborn as sns
iris= pd.DataFrame(x_train, columns=iris_dataset.feature_names)
# create a scatter matrix from the dataframe, color by y_train   hist_kwds={'bins': 20}
scatter_matrix(iris, c=y_train, figsize=(15, 15), marker='o', s=60, alpha=.8)
```

# From the above scatter matrix,it is clear that the species of iris has specific patterns. Since there are clearly visible patterns in all the features, all the features have been used to train the model.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn=knn.fit(x_train,y_train)
```

```python
pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred))
```

# An accuracy of 100% has been achieved by using KNeighborsClassifier for building the model. This accuracy confirms that this model is trustworthy and can be used for predictions.
